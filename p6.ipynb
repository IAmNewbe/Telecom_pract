## Antenna dan Saluran Transmisi

### Percobaan 1 Pengujian Reciprocity theorem pola radiasi antena yagi uda folded dipole 3 elemen

```python
import pandas as pd

# Nama file Excel
p8_raw = 'p6_raw.xlsx'

# Baca file Excel ke dalam DataFrame
df_1 = pd.read_excel(p8_raw, sheet_name='Sheet1')
df_1.head()
```

```python
import numpy as np

def hitung_daya(arus):
  hasil = (20 * np.log10(arus*0.000001)) + (10 * np.log10(50)) + 30
  return hasil

df_1['Daya Penerima (dBm) '] = hitung_daya(df_1['Arus Penerima (µA)  '])

df_1.head()
```

### Pola radiasi antena yagi uda folded dipole 3 elemen

```python
import numpy as np
import matplotlib.pyplot as plt
import math 

sudut = np.arange(0, 370, 10)
sudut_rad = np.radians(sudut)

# Plot radiasi antena
plt.figure(figsize=(4, 3))
plt.polar(sudut_rad, df_1['Daya Penerima (dBm) '])
plt.title('Plot Radiasi Antena yagi uda')
plt.xlabel('Sudut (degrees)')
plt.ylabel('Daya (dBm)')
plt.grid(True)
plt.show()
```
