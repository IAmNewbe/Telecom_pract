```python
import pandas as pd

# Array contoh
photo_1 = [[15.5, -18.51, 14.2],
         [19.2, -18, 15.8]]

# Nama kolom dan indeks baris
columns = ['Arus', 'dBm', 'uW']

# Membuat DataFrame dari array dengan nama kolom dan indeks baris
df_1 = pd.DataFrame(photo_1, columns=columns)

# Menampilkan DataFrame
print(df_1)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(30, 8))
def plot_data(photo, xx, yy, datanya):
  title = "Grafik pengaruh Arus terhadap Daya pada Photoemitter " + str(photo)
  plt.figure(figsize=(12, 6))
  sns.lineplot(y = yy, x = xx, data = datanya)
  
  # Menampilkan plot
  plt.title(title)
  plt.xlabel('Arus (mA)')
  plt.ylabel('Daya (uW)')
  plt.show()

plot_data( 1, 'Arus', 'uW', df_1)
```

```python
import pandas as pd

# Data untuk setiap photoemitter
data_photoemitter = {
    "Photoemitter#1": {
        "Current (mA)": 16.5,
        "Voltage (V)": 4.64,
        "Max Current (mA)": 19.1,
        "Max Voltage (V)": 4.71
    },
    "Photoemitter#2": {
        "Current (mA)": 16.1,
        "Voltage (V)": 4.201,
        "Max Current (mA)": 19.2,
        "Max Voltage (V)": 4.54
    },
    "Photoemitter#3": {
        "Current (mA)": 11.7,
        "Voltage (V)": 3.902,
        "Max Current (mA)": 13.7,
        "Max Voltage (V)": 4.253
    },
    "Photoemitter#4": {
        "Current (mA)": 16.5,
        "Voltage (V)": 3.796,
        "Max Current (mA)": 20,
        "Max Voltage (V)": 4.259
    },
    "Photoemitter#5": {
        "Current (mA)": 16.7,
        "Voltage (V)": 3.158,
        "Max Current (mA)": 20,
        "Max Voltage (V)": 3.332
    }
}

# Membuat dictionary untuk menyimpan dataframe untuk setiap photoemitter
dfs = {}
for photoemitter, data in data_photoemitter.items():
    # Menggabungkan kolom Max Current dan Max Voltage menjadi satu kolom
    data["Current (mA)"] = (data["Current (mA)"], data["Max Current (mA)"])
    data["Voltage (V)"] = (data["Voltage (V)"], data["Max Voltage (V)"])
    # Menghapus kolom Max Current dan Max Voltage yang tidak diperlukan lagi
    data.pop("Max Current (mA)")
    data.pop("Max Voltage (V)")
    df = pd.DataFrame(data)
    df['Daya (uW)'] = df['Voltage (V)'] * df['Current (mA)'] * 1000 #ubah ke mikro watt
    dfs[photoemitter] = df

# Menampilkan dataframe untuk setiap photoemitter
for photoemitter, df in dfs.items():
    print(f"Data untuk {photoemitter}:")
    print(df)
    print("\n")

```

```python
import matplotlib.pyplot as plt

def perbandingan(data1, data2):
  # Data
  x1 = data1['Arus']
  x2 = dfs[str(data2)]['Current (mA)']
  y1 = data1['uW']
  y2 = dfs[str(data2)]['Daya (uW)']

  # Membuat figure dan axes
  plt.figure(figsize=(8, 6))

  # Plot pertama
  plt.plot(x1, y1, color='blue', label='Pengukuran')

  # Plot kedua (menumpuk di atas plot pertama)
  plt.plot(x2, y2, color='red', label='Perhitungan')

  # Memberi judul dan label sumbu
  title = (f"Perbandingan Pengukuran dan Perhitungan daya {str(data2)}")
  plt.title(title)
  plt.xlabel('Arus (mA)')
  plt.ylabel('Daya (uW)')

  # Menambahkan legenda
  plt.legend()

  # Menampilkan plot
  plt.grid(True)
  plt.show()

perbandingan(df_1, "Photoemitter#1")
```
