```python
import numpy as np
import matplotlib.pyplot as plt

arus = [60.8, 60.3, 53, 53, 47, 39, 28, 16, 10, 3, 17, 13, 20, 26, 32, 35, 37, 42, 43.9 , 43, 42, 40, 36, 33, 25, 18, 10, 5, 4.5, 11.5, 20, 30, 36, 44, 50, 57, 60 ]
daya_antena = []
for nilai in arus:
  daya_antena.append((20 * math.log10(nilai*0.000001)) + (20 * math.log10(50))+ 30)
print(daya_antena)
print(len(daya_antena))
# Data sudut dalam derajat
sudut = np.arange(0, 370, 10)
print(len(sudut))
# Data daya antena (misalnya dalam dB)
# daya_antena = [20, 25, 30, 35, 40, 45, 50, 45, 40, 35, 30, 25, 20,10, 20, 25, 30, 35, 40, 45, 50, 45, 40, 35, ] 
print(len(daya_antena))
# Ubah sudut ke radian
sudut_rad = np.radians(sudut)

# Konversi daya dari dB ke faktor
daya_faktor = 10**(np.array(daya_antena)/10)


# Plot radiasi antena
plt.figure(figsize=(4, 3))
plt.polar(sudut_rad, daya_antena)
plt.title('Plot Radiasi Antena folded dipole')
plt.xlabel('Sudut (degrees)')
plt.ylabel('Daya (dBm)')
plt.grid(True)
plt.show()
```

### F/B Ratio

```python
front_1 = data_p1.loc[data_p1['Sudut (derajat)'] <= 90]
front_1 = pd.DataFrame(front_1)
front_2 = data_p1.loc[data_p1['Sudut (derajat)'] >= 270]  
front_2 = pd.DataFrame(front_2)

Front_Lobe_1 = pd.concat([front_1, front_2])
Front_Lobe_1_Max = Front_Lobe_1['Daya Penerima(dBm)'].max()
print("Front Lobe : ", Front_Lobe_1_Max)

back_1 = data_p1[(data_p1['Sudut (derajat)'] >= 90) & (data_p1['Sudut (derajat)']<= 270)]

Back_Lobe_1 = pd.DataFrame(back_1)
Back_Lobe_1_Max = Back_Lobe_1['Daya Penerima(dBm)'].max()
print("Back Lobe : ", Back_Lobe_1_Max)

print("folded dipole F/B Ratio : ", Front_Lobe_1_Max/Back_Lobe_1_Max)
```
