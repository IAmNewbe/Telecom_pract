### Redaman Fiber Optik

```python
import matplotlib.pyplot as plt

redaman_fiber_optik = [0.15, 0.21, 0.06, 0.98]
panjang_gelombang = [526, 590, 660, 850]

# Plot garis
def plot_garis(judul, arus, daya):
  plt.plot(arus, daya, marker='o', linestyle='-')

  # Menambahkan label dan judul
  plt.xlabel('Panjang Gelombang (nm)')
  plt.ylabel('redaman/m')
  plt.title(judul)

  # Menampilkan plot
  plt.grid(True)
  plt.show()

plot_garis("hubungan panjang gelombang dengan redaman/m", panjang_gelombang, redaman_fiber_optik)

```
